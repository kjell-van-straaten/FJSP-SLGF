```python
# imports
import os
import sys
sys.path.append('../../FJSP-SLGF/')
import pandas as pd
import dataset_loader
import dataset_loader_sdst
from pprint import pprint
from utils import WfInstance
import pickle
import numpy as np
```

FJSP

```python
root_path = "C:\git-repos\FJSP-SLGF\datasets\FJSP"

results = {}

for instance_type in ['mkdata', 'edata',  'rdata', 'vdata', 'sodata']:

    total_operations = []
    total_jobs = []
    total_machines = []
    operation_options = []
    processing_times = []
    folder_path = os.path.join(root_path, instance_type)
    nr_instances = 0
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".fjs"):
                nr_instances += 1
                instance_path = os.path.join(root, file)
                dataset = dataset_loader.load_lit_dataset(instance_path)
                total_operations.append(dataset.nr_operations)
                total_jobs.append(dataset.nr_jobs)
                total_machines.append(dataset.nr_machines)
                options = sum([sum(i) for i in dataset.ope_machines]) / len(dataset.ope_machines)
                operation_options.append(options)
                processing_times.append(sum([sum(i) for i in dataset.duration]) / (options * dataset.nr_operations))
            results[instance_type] = {
                "Number of Instances": nr_instances,
                "Average Number of Jobs": sum(total_jobs) / len(total_jobs),
                "Max Number of Jobs": max(total_jobs),
                "Min Number of Jobs": min(total_jobs),
                "Average Number of Machines": sum(total_machines) / len(total_machines),
                "Max Number of Machines": max(total_machines),
                "Min Number of Machines": min(total_machines),
                "Average Number of Operations": sum(total_operations) / len(total_operations),
                "Max Number of Operations": max(total_operations),
                "Min Number of Operations": min(total_operations),
                "Average Operation Options": sum(operation_options) / len(operation_options),
                "Average Processing Time": sum(processing_times) / len(processing_times),
            }
    
df = pd.DataFrame(results)
df.round(1)
```

```python
df.round(1).to_latex()
```

FJSP SDST


```python
root_path = "C:\git-repos\FJSP-SLGF\datasets\FJSP_SDST"

results = {}

for instance_type in ['ftdata', 'sodata']:

    total_operations = []
    total_jobs = []
    total_machines = []
    operation_options = []
    processing_times = []
    sdst = []
    folder_path = os.path.join(root_path, instance_type)
    nr_instances = 0
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".fjs") or file.endswith(".pickle"):
                nr_instances += 1
                instance_path = os.path.join(root, file)
            else:
                continue
            if file.endswith(".fjs"):
                dataset = dataset_loader_sdst.load_lit_dataset_sdst(instance_path)
            else:
                with open (instance_path, 'rb') as handle:
                    dataset: WfInstance = pickle.load(handle)
            total_operations.append(dataset.nr_operations)
            total_jobs.append(dataset.nr_jobs)
            total_machines.append(dataset.nr_machines)
            options = sum([sum(i) for i in dataset.ope_machines]) / len(dataset.ope_machines)
            operation_options.append(options)
            processing_times.append(sum([sum(i) for i in dataset.duration]) / (options * dataset.nr_operations))
            sdsts = np.concatenate(np.array(dataset.sdst)).ravel().tolist()
            sdsts = [i for i in sdsts if i != 1000000]
            [sdst.append(j) for j in sdsts]



    results[instance_type] = {
        "Number of Instances": nr_instances,
        "Average Number of Jobs": sum(total_jobs) / len(total_jobs),
        "Max Number of Jobs": max(total_jobs),
        "Min Number of Jobs": min(total_jobs),
        "Average Number of Machines": sum(total_machines) / len(total_machines),
        "Max Number of Machines": max(total_machines),
        "Min Number of Machines": min(total_machines),
        "Average Number of Operations": sum(total_operations) / len(total_operations),
        "Max Number of Operations": max(total_operations),
        "Min Number of Operations": min(total_operations),
        "Average Operation Options": sum(operation_options) / len(operation_options),
        "Average Processing Time": sum(processing_times) / len(processing_times),
        "Average SDST": sum(sdst) / len(sdst),
        "Max SDST": max(sdst),
        "Min SDST": min(sdst)

    }

df = pd.DataFrame(results)
df.round(1)
```

FJSP SRMN

```python
root_path = "C:\git-repos\FJSP-SLGF\datasets\FJSP_SRMN"

results = {}

for instance_type in ['wf_data']:

    total_operations = []
    total_jobs = []
    total_machines = []
    operation_options = []
    processing_times = []
    sdst = []
    fraction = []
    release_times = []
    fraction_operation_night_setup  = []
    fraction_operation_night_operation = []
    folder_path = os.path.join(root_path, instance_type)
    nr_instances = 0
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".fjs") or file.endswith(".pickle"):
                nr_instances += 1
                instance_path = os.path.join(root, file)
            else:
                continue
            if file.endswith(".fjs"):
                dataset = dataset_loader_sdst.load_lit_dataset_sdst(instance_path)
            else:
                with open (instance_path, 'rb') as handle:
                    dataset: WfInstance = pickle.load(handle)
            total_operations.append(dataset.nr_operations)
            total_jobs.append(dataset.nr_jobs)
            total_machines.append(dataset.nr_machines)
            options = sum([sum(i) for i in dataset.ope_machines]) / len(dataset.ope_machines)
            operation_options.append(options)
            processing_times.append(sum([sum(i) for i in dataset.duration]) / (options * dataset.nr_operations))
            sdsts = np.concatenate(np.array(dataset.sdst)).ravel().tolist()
            sdsts = [i for i in sdsts if i != 0]
            [sdst.append(j) for j in sdsts]
            fraction.append((np.array(dataset.job_release_dates)> 0).sum() /len(dataset.job_release_dates))
            release_times.append(sum(dataset.job_release_dates) / (np.array(dataset.job_release_dates)> 0).sum())
            fraction_operation_night_setup.append(sum(dataset.ope_night_setup) / len(dataset.ope_night_setup))
            fraction_operation_night_operation.append(sum(dataset.ope_night_operation) / len(dataset.ope_night_operation))

    #replace nan valeus with 0
    release_times = [0 if np.isnan(i) else i for i in release_times]
    results[instance_type] = {
        "Number of Instances": nr_instances,
        "Average Number of Jobs": sum(total_jobs) / len(total_jobs),
        "Max Number of Jobs": max(total_jobs),
        "Min Number of Jobs": min(total_jobs),
        "Average Number of Machines": sum(total_machines) / len(total_machines),
        "Max Number of Machines": max(total_machines),
        "Min Number of Machines": min(total_machines),
        "Average Number of Operations": sum(total_operations) / len(total_operations),
        "Max Number of Operations": max(total_operations),
        "Min Number of Operations": min(total_operations),
        "Average Operation Options": sum(operation_options) / len(operation_options),
        "Average Processing Time": sum(processing_times) / len(processing_times),
        "Average SDST": sum(sdst) / len(sdst),
        "Max SDST": max(sdst),
        "Min SDST": min(sdst),
        "Average Release Time": sum(release_times) / len(release_times),
        "Average Fraction of Jobs with Release Time": sum(fraction) / len(fraction),
        "Fraction Operation Night Operation": sum(fraction_operation_night_operation) / len(fraction_operation_night_operation),
        "Fraction Operation Night Setup": sum(fraction_operation_night_setup) / len(fraction_operation_night_setup),

    }

df = pd.DataFrame(results)
df.round(1)
```
